# Streaming v/s Batch Response

Normally when used invoke() the response is generated at once i.e. the output will be displayed when it is completely generated <br>
else where in Streaming the response is displayed simultenously as it is generated stream() is used<br>
and in batch we can execute mulitple requests at a time and it will be executed parallely.

```python
# Streaming Example
for chunk in model.stream("Write a 20 words essay on AI"):
    print(chunk.text, end="-", flush=True)
```

```python
# Batch Example
responses = model.batch(
    [
        "Why do parrots have colorful feathers?",
        "How do airplanes fly?",
        "What is quantum computing?"
    ]
)
for response in responses:
    print(response)
```
